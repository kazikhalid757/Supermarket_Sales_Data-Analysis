```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("supermarket_sales_Dataset.csv")
df.head()
```

### Which city sells more?

```python

# Count the frequency of each city
city_counts = df['City'].value_counts()

# Create the pie chart
plt.pie(city_counts, labels=city_counts.index,autopct='%1.2f%%')

# Add a title
plt.title('Sales by City')

# Show the chart
plt.show()
```

### Which Branch sells more?

```python
branch_counts=df['Branch'].value_counts()

plt.pie(branch_counts,labels=branch_counts.index,autopct='%1.2f%%')

# Add a title
plt.title('Sales by Brance')

# Show the chart
plt.show()
```

### Which Customer type buy more?

```python
customer_c=df['Customer type'].value_counts()
plt.pie(customer_c,labels=customer_c.index,autopct='%1.2f%%')
plt.title('Which Customer type buy more?')
plt.show
```

### Which Customer buy more?

```python
Gender_g=df['Gender'].value_counts()
plt.pie(Gender_g,labels=Gender_g.index,autopct='%1.2f%%')
plt.title('Which Customer buy more?')
plt.show
```

### Which Payment Method use more?

```python
Payment_p=df['Payment'].value_counts()
plt.pie(Payment_p,labels=Payment_p.index,autopct='%1.2f%%')
plt.title('Which Payment Method use more?')
plt.show
```

### Which Product line  sell  more?

```python
grouped_df = df.groupby('Product line')['Quantity'].agg('sum')

# plot the grouped data
plt.figure(figsize=(15, 8))
plt.bar(grouped_df.index, grouped_df.values,width = 0.20)
plt.xlabel('Product Line')
plt.ylabel('Quantity')
plt.title('Quantity per Product Line')
plt.show()
```

```python
female_df = df[df['Gender'] == 'Female']
grouped_df = female_df.groupby('Product line')['Quantity'].sum()

plt.figure(figsize=(15, 8))
plt.bar(grouped_df.index, grouped_df.values, width=0.4)
plt.xlabel('Product Line')
plt.ylabel('Quantity')
plt.title('Quantity per Product Line (Female Customers)')
plt.show()
```

```python
female_df = df[df['Gender'] == 'Male']
grouped_df = female_df.groupby('Product line')['Quantity'].sum()

plt.figure(figsize=(15, 8))
plt.bar(grouped_df.index, grouped_df.values, width=0.4)
plt.xlabel('Product Line')
plt.ylabel('Quantity')
plt.title('Quantity per Product Line (Male Customers)')
plt.show()
```
